### 서울시 부동산 실거래가

### 연립다세대(전월세)

```python
import pandas as pd

df = pd.read_csv('./data/연립다세대(전월세)_실거래가_20250924171625.csv', header=15 , encoding='cp949')
df2 = pd.read_csv('./data/연립다세대(매매)_실거래가_20250924172243.csv', header=15 , encoding='cp949')

df = pd.concat([df, df2])
```

```python
df.columns
```

```python
df['보증금(만원)'].fillna('0', inplace=True)
df['보증금(만원)'] = df['보증금(만원)'].str.replace(',', '').astype(int)
```

#### 전세만

```python
cond1 = df['전월세구분'] == '전세'
cond2 = df['보증금(만원)']<30000
cond3 = df['전용면적(㎡)']>50
cond4 = df['건축년도'] > 2000

df = df[cond1 & cond2 & cond3 & cond4]
```

```python
df.columns
```

```python
df = df[['시군구', '건물명', '전월세구분', '전용면적(㎡)', '계약년월', '계약일', '보증금(만원)', '층', '건축년도', '계약기간', '계약구분', '종전계약 보증금(만원)']]
```

```python
df['구'] = df['시군구'].str.split(' ').str[1]
```
